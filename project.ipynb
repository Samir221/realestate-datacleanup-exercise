#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# Define the populations in the "south belt" of Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create the subset DataFrame
south_belt_df = ds[ds['level5'].isin(south_belt_populations)]

# Normalize prices using Min-Max scaling for each population
south_belt_df['normalized_price'] = south_belt_df.groupby('level5')['price'].transform(
    lambda x: (x - x.min()) / (x.max() - x.min())
)

# Plot histograms of normalized prices
plt.figure(figsize=(12, 8))
populations = south_belt_populations

for population in populations:
    subset = south_belt_df[south_belt_df['level5'] == population]
    plt.hist(subset['normalized_price'], bins=10, alpha=0.5, label=population, edgecolor='black')

plt.title('Histograms of Normalized Prices for Each Population in the South Belt of Madrid')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# Define the populations in the "south belt" of Madrid
south_belt_populations = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Create the subset DataFrame
south_belt_df = ds[ds['level5'].isin(south_belt_populations)]

# Create a new column for price per square meter
south_belt_df['pps'] = south_belt_df['price'] / south_belt_df['surface']

# Plot scatter plots of price per square meter for each population using subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharey=True)

populations = south_belt_populations

for ax, population in zip(axes.flatten(), populations):
    subset = south_belt_df[south_belt_df['level5'] == population]
    ax.scatter(subset['surface'], subset['pps'], alpha=0.7, edgecolors='w', linewidth=0.5)
    ax.set_title(f'{population}')
    ax.set_xlabel('Surface Area (m²)')
    ax.set_ylabel('Price per Square Meter (USD/m²)')
    ax.grid(True)

fig.suptitle('Price per Square Meter vs. Surface Area for Different Populations in the South Belt of Madrid', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
